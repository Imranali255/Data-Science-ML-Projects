# pyplot vs Object Oriented Method

```python
heart = pd.read_csv('heart_disease_uci.csv')
heart.dropna(inplace=True)
heart
```

```python
heart.head()
```

```python
over_40 = heart[heart['age']>40]
over_40
```

```python
over_40.plot(kind='scatter',
            x='age',
            y='chol',
            c='num');
```

```python
# object Oriented method
fig, ax= plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',
            x='age',
            y='chol',
            c='num',
            ax=ax);
```

# Entire Life Cycle of Object Oriented Plot

```python
# creat figure and graph area
fig,ax = plt.subplots(figsize=(10,6))
#home, room
# plot the data
scatter_plot = ax.scatter(x=over_40['age'],
                         y=over_40['chol'],
                         c=over_40['num'])
# customization
ax.set(title='Suspectability of Heart Disease',
      xlabel='Age',
      ylabel='Cholesterol')

# set the Legend
ax.legend(*scatter_plot.legend_elements(),title='num')

scatter_plot;
```

#  Entire Life Cycle of Object Oriented Plot Advanced 

```python
#1 create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,
                                 ncols=1,
                                 figsize=(10,10),
                                sharex=True)
#2 add data to plot 1
scatter_plot = plot1.scatter(x=over_40['age'],
                            y=over_40['chol'],
                            c=over_40['exang'],
                            cmap='winter')
#3 customize
plot1.set(title='Heart Disease Problem with Cholesterol',
         #xlabel='Age',
         ylabel='Cholesterol')
plot1.legend(*scatter_plot.legend_elements(),title='Survived or Not')
plot1.axhline(y=over_40['chol'].mean(),linestyle='--')

###### Now Plot Second Plot
#4 add data to plot 2

scatter_plot = plot2.scatter(x=over_40['age'],
                            y=over_40['trestbps'],
                            c=over_40['exang'],
                            cmap='GnBu')
#5 Customize
plot2.set(title='Heart Disease problem with Resting Blood Pressure',
         xlabel='Age',
         ylabel='Resting Blood Pressure')
plot2.legend(*scatter_plot.legend_elements(),title='Survived or Not');
#6 Meanline
plot2.axhline(y=over_40['trestbps'].mean(),linestyle='--')
#7 coloring data cmap='GnBu')
#8 coloring plots
plt.style.use('seaborn-v0_8-darkgrid')
#9 Naming the entire figure
fig.suptitle('Patients Data',fontsize=32,fontweight='bold');
#10 saving the figure
fig.savefig('patients data report.png')
```
