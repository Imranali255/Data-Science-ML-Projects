## Classfication Problem

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(7)
# Loading Data
heart_disease = pd.read_csv('heart_disease_uci.csv')
heart_disease.dropna(inplace=True)   # NAN value removed 
# Spliting Data
X = heart_disease.drop('num', axis=1)
y = heart_disease['num']
# categorical error solved
categorical_columns = ['sex', 'dataset', 'cp', 'restecg', 'slope', 'thal']  
le = LabelEncoder()
for col in categorical_columns:
   X[col] = le.fit_transform(X[col])
# Make test and train sets
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size =0.20)
# Instatiate the model
clf = RandomForestClassifier(n_jobs=-1)
# Fit the model
clf.fit(X_train,y_train)
# Lets Make Predictions
y_predicted =clf.predict(X_test)
# Evaluating the Model using sklearn Functions
print("Model Evaluation Using Sklearn Functions")
print(f':Acuuracy_Score:{accuracy_score(y_test,y_predicted)}')
print(f':Precision_score:{precision_score(y_test,y_predicted,zero_division=1,average='macro')}')
print(f':Recall_scor:{recall_score(y_test,y_predicted,average='macro')}')
print(f':F1_score:{f1_score(y_test,y_predicted,average='macro')}')
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
def evaluate_hyper_metrics(y_true,y_predicted):
    """This function return the Evaluation Metrics""" 
    accuracy = accuracy_score(y_true,y_predicted)
    precision = precision_score(y_true,y_predicted)
    recall = recall_score(y_true,y_predicted)
    f1 = f1_score(y_true,y_predicted)

    metrics_dict = {"Accuracy": accuracy,
                   "precision":precision,
                   "recall":recall,
                   "f1":f1}
    print(f"Accuracy:{round(accuracy,2)}")
    print(f"precision:{round(precision,2)}")
    print(f"recall:{round(recall,2)}")
    print(f"f1:{round(f1,2)}")
    
    return metrics_dict
```

```python
# Train
# Validate
# Test
mixed_heart_disease = heart_disease.sample(frac=1)
# Split X,y
X = mixed_heart_disease.drop('target' , axis=1)
y = mixed_heart_disease['target']
#splitting data into three parts Now: Train =70%,Validate=15%,Test=15%
train_split = round(0.7*len(mixed_heart_disease))
valid_split = round(train_split + 0.15*len(mixed_heart_disease))

X_train,y_train = X[:train_split],y[:train_split]
X_valid,y_valid = X[train_split:valid_split],y[train_split:valid_split]
X_test,y_test = X[valid_split:],y[valid_split:]
```

```python
np.random.seed(7)
clf = RandomForestClassifier(max_depth=2, min_samples_split=20, random_state=42)
# max_depth=2, - Limits the depth of each decision tree in the forest to 2 levels.prevent overfitting but may reduce model complexity.
# min_samples_split=20: Ensures that a node must have at least 20 samples before it can be split into child nodes.
#random_state=42: Sets a fixed seed for randomness to ensure reproducibility of results. Choosing 42 is common among data scientists for consistency.

clf.fit(X_train,y_train)
# Baseline Prediction
y_predicted = clf.predict(X_valid)
# Evaluation the model
first_result = evaluate_hyper_metrics(y_valid,y_predicted)
```

```python
# Second Model results  
np.random.seed(7)
clf2 = RandomForestClassifier(n_estimators=10)
clf2.fit(X_train,y_train)
# Baseline Prediction
y_predicted = clf2.predict(X_valid)
# Evaluation the model
second_result = evaluate_hyper_metrics(y_valid,y_predicted)
```

```python
# Third Model results  
np.random.seed(7)
clf3 = RandomForestClassifier(n_estimators=10,max_depth=10)
clf3.fit(X_train,y_train)
# Baseline Prediction
y_predicted = clf3.predict(X_valid)
# Evaluation the model
third_result = evaluate_hyper_metrics(y_valid,y_predicted)
```

```python
hparams = {"max_depth":[None,5,10,20,30],
           "n_estimators":[10,100,200,500],
           "min_samples_split":[2,4,6],
           "min_samples_leaf":[1,2,4],
           "max_features":["sqrt","log2"]}
# Split  into X and y
X = mixed_heart_disease.drop("target",axis=1)
y = mixed_heart_disease["target"]
# Split into train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

clf = RandomForestClassifier(n_jobs=-1)
```

### Fill the missing Value
### Convert the data type object type to into num type
### Build a Model on the Data

```python
# Getting data ready
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
# Modeling
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split,GridSearchCV
np.random.seed(7)

# Import Data and filll the missing Values

data = pd.read_csv("smartphones.csv")
data['Final Price'] = data['Final Price'].str.replace(r'[\$\,\.]','',regex=True)
data['Final Price'] = data['Final Price'].str[:3].astype(int)
data.dropna(subset = ["Final Price"], inplace=True)   # when there is nan value than use this otherwise not use this

# Defibe Different features  and Transformer pipeline
cat_features = ['Smartphone','Brand','Model','Color','Free']

########################11111#######################
cat_transformer = Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="constant",fill_value="missing")),
    ("onehot",OneHotEncoder(handle_unknown="ignore"))
])
########################2222#######################
Ram_features =["RAM"]
Ram_transformer = Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="constant",fill_value=4))

])
#######################3333#######################
num_feature = ["Storage"]
num_transformer = Pipeline(steps=[
    ("imputer",SimpleImputer(strategy="mean"))
])
################################################
# step preprocessing step (fill the missing values, then convert to numbers)

preprocessing = ColumnTransformer(
    transformers=[
        ("cat",cat_transformer,cat_features),
        ("Ram",Ram_transformer,Ram_features),
        ("num",num_transformer,num_feature)
    ]
)



model = Pipeline(steps=[
    ("preprocessing",preprocessing),
    ("model",RandomForestRegressor())
])

#split Data

X =  data.drop("Final Price",axis=1)
y = data["Final Price"]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# now run the fit and evaluates
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# Tuning the hyper parameteters using GridSearchCV using pipeline
hparams3 = {
    "preprocessing__num__imputer__strategy":["mean","median"],
    "model__max_depth": [10,20],
    "model__n_estimators": [100,200],
    "model__min_samples_split": [2,4],
    "model__min_samples_leaf": [1,2],
}

# Initialize GridSearchCV
my_gscv_model = GridSearchCV(model, hparams3, cv=5, verbose=2,n_jobs=-1)

# Fit GridSearchCV
my_gscv_model.fit(X_train, y_train)
```

```python
my_gscv_model.score(X_test,y_test)
```
